```python
import os
import shutil
import random
from tqdm.notebook import tqdm
```

```python
!pwd
```

```python
train_img_path='./yolo_data/images/train/'
train_label_path='./yolo_data/labels/train/'
val_img_path='./yolo_data/images/val/'
val_label_path='./yolo_data/labels/val/'
test_path='./yolo_data/test/'
```

```python
random.seed(42)
```

```python
def train_testsplit(path, split=0.2):
    file = list(set([name[:-4] for name in os.listdir(path)]))
    print(file)
    random.shuffle(file)

    # train test split
    test_size = int(len(file) * split)
    print("test images : ", test_size)
    train_size = len(file) - test_size
    print("train names : ", train_size)

    # create directories
    os.makedirs(train_img_path, exist_ok=True)
    os.makedirs(train_label_path, exist_ok=True)
    os.makedirs(val_img_path, exist_ok=True)
    os.makedirs(val_label_path, exist_ok=True)

    # training area
    for fname in tqdm(file[:train_size], desc="Copying train data"):
        shutil.copy2(path + fname + '.jpg', f"{train_img_path}/" + fname + '.jpg')
        shutil.copy2(path + fname + '.txt', f"{train_label_path}/" + fname + '.txt')

    # validation/testing
    for fname in tqdm(file[train_size:], desc="Copying val data"):
        shutil.copy2(path + fname + '.jpg', f"{val_img_path}/" + fname + '.jpg')
        shutil.copy2(path + fname + '.txt', f"{val_label_path}/" + fname + '.txt')
```
